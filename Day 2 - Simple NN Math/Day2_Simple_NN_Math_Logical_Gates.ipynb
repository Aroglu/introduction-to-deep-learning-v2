## Conventional Python Solutions To Logical Gates

```python
def LogicalNOT(x1):
  return not x1

def LogicalAND(x1,x2):
  return x1 and x2

def LogicalNAND(x1, x2):
  return not (x1 and x2)

def LogicalOR(x1, x2):
  return x1 or x2

def LogicalNOR(x1, x2):
  return not (x1 or x2)

def LogicalXOR(x1, x2):
  return (not x1 and x2) or (x1 and not x2)

def LogicalXNOR(x1, x2):
  return (x1 and x2) or (not x1 and not x2)
```

```python
#NOT
print(LogicalNOT(0))
print(LogicalNOT(1))
```

```python
#AND
print(LogicalAND(0,0))
print(LogicalAND(0,1))
print(LogicalAND(1,0))
print(LogicalAND(1,1))
```

```python
#NAND
print(LogicalNAND(0,0))
print(LogicalNAND(0,1))
print(LogicalNAND(1,0))
print(LogicalNAND(1,1))
```

```python
#OR
print(LogicalOR(0,0))
print(LogicalOR(0,1))
print(LogicalOR(1,0))
print(LogicalOR(1,1))
```

```python
#NOR
print(LogicalNOR(0,0))
print(LogicalNOR(0,1))
print(LogicalNOR(1,0))
print(LogicalNOR(1,1))
```

```python
#XOR
print(LogicalXOR(0,0))
print(LogicalXOR(0,1))
print(LogicalXOR(1,0))
print(LogicalXOR(1,1))
```

```python
#XNOR
print(LogicalXNOR(0,0))
print(LogicalXNOR(0,1))
print(LogicalXNOR(1,0))
print(LogicalXNOR(1,1))
```

```python
#import required modules
from tensorflow import keras
import numpy as np
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
def ValidateGateModel(model, input, output):
  predictions = model.predict(input).round()
  
  for idx in range(len(predictions)):
    print(f"Round off Prediction Value - Actual Value : {predictions[idx]} - {output[idx]}")
```

```python
#NOT
def TrainValidateNOTGateModel():
  input = np.array([[1],[0]])
  output = np.array([[0],[1]])

  model = Sequential()
  model.add(Dense(32, input_dim=1)) # If you don't assign any activation function 'linear' activation function will be applied
  model.add(Dense(1)) # Output layer has 1 neuron

  model.compile(loss='mean_squared_error',optimizer='adam', metrics=['binary_accuracy'])

  model.fit(input, output, epochs=100, verbose=0)

  ValidateGateModel(model, input, output)
```

```python
#AND
def TrainValidateANDGateModel():
  input = np.array([[0,0],[0,1],[1,0],[1,1]])
  output = np.array([[0],[0],[0],[1]])

  model = Sequential()
  model.add(Dense(16, input_dim=2, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))

  model.compile(loss='mean_squared_error',optimizer='adam', metrics=['binary_accuracy'])

  model.fit(input, output, epochs=600, verbose=0)

  ValidateGateModel(model, input, output)
```

```python
#OR
def TrainValidateORGateModel():
  input = np.array([[0,0],[0,1],[1,0],[1,1]])
  output = np.array([[0],[1],[1],[1]])

  model = Sequential()
  model.add(Dense(16, input_dim=2, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))

  model.compile(loss='mean_squared_error',optimizer='adam', metrics=['binary_accuracy'])

  model.fit(input, output, epochs=600, verbose=0)

  ValidateGateModel(model, input, output)
```

```python
#XOR
def TrainValidateXORGateModel():
  input = np.array([[0,0],[0,1],[1,0],[1,1]])
  output = np.array([[0],[1],[1],[1]])

  model = Sequential()
  model.add(Dense(16, input_dim=2, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))

  model.compile(loss='mean_squared_error',optimizer='adam', metrics=['binary_accuracy'])

  model.fit(input, output, epochs=6000, verbose=0)

  ValidateGateModel(model, input, output)
```

```python
TrainValidateNOTGateModel()
```
