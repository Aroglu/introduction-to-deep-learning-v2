```python
from tensorflow import keras
from keras import layers, models, losses
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.preprocessing import image
```

```python
#import data
(train_images,train_labels),(test_images,test_labels) = keras.datasets.cifar10.load_data()
```

```python
class_names=["airplane","automobile","bird","cat",
             "deer","dog","frog","horse","ship","truck"]
```

```python
#scaling
train_images, test_images = train_images / 255, test_images / 255
```

```python
#Create model
model = models.Sequential()
```

```python
#Create CNN Layers
model.add(layers.Conv2D(32,(3,3),activation="relu",input_shape=(32,32,3)))
model.add(layers.MaxPooling2D())
model.add(layers.Conv2D(64,(3,3),activation="relu"))
model.add(layers.MaxPooling2D())
model.add(layers.Conv2D(64,(3,3),activation="relu"))
```

```python
#Create Dense Layers
model.add(layers.Flatten())
model.add(layers.Dense(64,activation="relu"))
model.add(layers.Dense(10))
```

```python
model.summary()
```

```python
#Compile Model
model.compile(optimizer="adam",
              loss = losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=["accuracy"])
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
# To increase our input data we use Image Augmentation
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2
)
```

```python
# Fit the train_images to datagen
datagen.fit(train_images)
```

```python
#Fit the model with datagen
history = model.fit(datagen.flow(train_images, train_labels, subset="training"), 
          batch_size = 1024, steps_per_epoch=len(train_images)/1024, validation_data = datagen.flow(train_images, train_labels, subset="validation"), 
          epochs=10, verbose=1)
```

```python
#evaluate the model and get loss with accuracy metric
model.evaluate(datagen.flow(test_images, test_labels),batch_size=1024)
```

```python
# list all data in history
print(history.history.keys())
```

```python
# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history["val_loss"])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
